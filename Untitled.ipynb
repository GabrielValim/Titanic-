## Tratando valores vazios

```python
def extrai_titulo(texto):
    titulos = texto[texto.find(",")+2:texto.find(".")]
    
    if titulos in ["Mr","Miss","Mrs","Master"]:
        return titulos
    else:
        return "outros"
```

```python
train_df["Titulos"] = train_df.Name.apply(extrai_titulo)
test_df["Titulos"] = test_df.Name.apply(extrai_titulo)
```

```python
#substituindo valores nulos pelo media da idade com relacao a classe sexo e titulos
train_df["Age"] = train_df.Age.fillna(train_df.groupby(["Pclass","Sex","Titulos"])["Age"].transform("mean"))
test_df["Age"] = train_df.Age.fillna(test_df.groupby(["Pclass","Sex","Titulos"])["Age"].transform("mean"))
```

```python
train_df.loc[train_df.Embarked.isnull(),"Embarked"] = "S"
test_df.loc[test_df.Embarked.isnull(),"Embarked"] = "S"
```

```python
#valores com alta cardinaliedade
train_df = train_df.drop("Cabin",axis=1)
test_df = test_df.drop("Cabin",axis=1)
```

```python
train_df = train_df.drop(["PassengerId","Name","Ticket"], axis=1)
test_df = test_df.drop(["PassengerId","Name","Ticket"], axis=1)
```

## Feature engineering

```python
# viabilizando dados
train_df.info()
train_df.head()
```

```python
#transformando a coluna sex
train_df["Is_male"] = train_df.Sex.apply(lambda x: 1 if x == "male" else 0)
test_df["Is_male"] = test_df.Sex.apply(lambda x: 1 if x == "male" else 0)
```

```python
#transformando a coluna embarked e titulos
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe = ohe.fit(train_df.drop("Survived",axis=1)[["Embarked","Titulos"]])
ohe_df = pd.DataFrame(ohe.transform(train_df.drop("Survived",axis=1)[["Embarked","Titulos"]]).toarray())
ohe_df.columns = ohe.get_feature_names_out()
train_df = pd.concat([train_df,ohe_df],axis=1)
```

```python
ohe = ohe.fit(test_df[["Embarked","Titulos"]])
ohe_df = pd.DataFrame(ohe.transform(test_df[["Embarked","Titulos"]]).toarray())
ohe_df.columns = ohe.get_feature_names_out()
test_df = pd.concat([test_df,ohe_df],axis=1)
```

```python
train_df = train_df.drop(["Embarked","Titulos","Sex"],axis=1)
test_df = test_df.drop(["Embarked","Titulos","Sex"],axis=1)
```

## Padronizaçao e Normalizaçao

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(train_df[["Age"]])
train_df["Age_padrao"] = scaler.transform(train_df[["Age"]])
```

```python
scaler = StandardScaler().fit(test_df[["Age"]])
test_df["Age_padrao"] = scaler.transform(test_df[["Age"]])
```

```python
scaler = StandardScaler().fit(train_df[["Fare"]])
train_df["Fare_padrao"] = scaler.transform(train_df[["Fare"]])
```

```python
scaler = StandardScaler().fit(test_df[["Fare"]])
test_df["Fare_padrao"] = scaler.transform(test_df[["Fare"]])
```

```python
train_df = train_df.drop(["Fare","Age"],axis=1)
test_df = test_df.drop(["Fare","Age"],axis=1)
```

```python
test_df.loc[test_df.Fare_padrao.isnull(),"Fare_padrao"] = 0
```

## Modelos

```python
X_train = train_df.drop("Survived", axis=1)
Y_train = train_df["Survived"]
X_test = test_df.copy()
```

```python
#Regressao logistica

logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
#knn
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

```python
# Perceptron

perceptron = Perceptron()
perceptron.fit(X_train, Y_train)
Y_pred = perceptron.predict(X_test)
acc_perceptron = round(perceptron.score(X_train, Y_train) * 100, 2)
acc_perceptron
```

```python
# Decision Tree

decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_train, Y_train) * 100, 2)
acc_decision_tree
```

```python
# Random Forest

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```

```python
modelos = pd.DataFrame({
    'Modelos': ['KNN', 'Logistic Regression', 
              'Random Forest','Perceptron', 
              'Decision Tree'],
    'Score': [acc_knn, acc_log, 
              acc_random_forest, acc_perceptron, 
              acc_decision_tree]})

modelos.sort_values(by='Score', ascending=False)
```
